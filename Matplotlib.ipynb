```python
# plt.plot(x,y, color= '',linewidth = , linestyle= '',label='')
import matplotlib.pyplot as plt
months = [1,2,3,4]
sales = [1000,500,20000,200]

plt.plot(months , sales , color = 'orange' , linestyle = '--', marker = 'o',label = 'Sales for a month')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.legend()
plt.grid(color='purple',linestyle=':',linewidth=1)
plt.xticks([1,2,3,4], ['M1','M2','M3','M4'])
plt.title('Monthy Sales data report')
plt.show()
```

```python
# plt.bar()  For showing the graphs like barchart
import matplotlib.pyplot as plt
product = ['A','B','C','D']
sales = [100,300,60,335]
plt.bar(product,sales,color='magenta',label='Sales 2025') # for vertical plt.barv() , for horizontal plt.barh()
plt.xlabel('Products')
plt.ylabel('sales')
plt.legend()
plt.title("Sales for 2025")
plt.show()
```

```python
# piechart
import matplotlib.pyplot as plt
import random

# assigning the names
names = ['Arjun','Krishna','Bhishma','Karn']

# Generation random number using randInt
scores = [random.randint(1,10) for name in names]

# plotting a piechart
plt.pie(scores,labels=names,colors=['blue','grey','gold','orange'],autopct ='%1.1f%%')
plt.title("Powers of God")
plt.show()
```

```python
# Making a small project , calulating the batting percentage of friends
import matplotlib.pyplot as plt
names = ['Harsh','Ritik','Aakash','Dubey']
hitting = [90,75,65,20]
plt.pie(hitting,labels=names,autopct='%1.1f%%')
plt.title("Power hitting ability of Players")
plt.show()
```

```python
# histogram
import matplotlib.pyplot as plt
students = [22,33,45,65,89,56,34,34,56,78,24,74,63,64]
plt.hist(students,bins=5,color='purple',edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('Number of Student')
plt.title('Range of score according to the Students')
plt.show()
```

```python
# Scatter Plot
import matplotlib.pyplot as plt
hours = [1,2,3,4,5,6]
exam_score = [55,66,97,45,88,69]

plt.scatter(hours,exam_score,color='green',marker='o',label='Marks Analysis')
plt.xlabel('Hours of Study by Student')
plt.ylabel('Score of Students')
plt.grid()
plt.title('Scatter plot of students marks according to the hour of study')
plt.show()
```

```python
# subplot
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [15,22,31,12]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar Chart')
plt.show()
```

```python
# fig , ax = plt.subplots(nrows,ncols,figsize=(width,height))
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,2, figsize = (10,5))
x = [1,2,3,4]
y = [15,22,31,12]
ax[0].plot(x,y)
ax[0].set_title('Line Chart')

ax[1].bar(x,y)
ax[1].set_title('Bar Chart')
plt.tight_layout()
plt.show()
```
