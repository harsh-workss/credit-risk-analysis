```python
# Using Normalization to range numbers btw 0 to 1 
import numpy as np
import pandas as pd

file_path = r"C:\Users\Harsh Yadav\.cache\kagglehub\datasets\rakeshkapilavai\extrovert-vs-introvert-behavior-data\versions\2\personality_datasert.csv"
df = pd.read_csv(file_path)

for col in df.select_dtypes(include=['float64','int64']).columns:
    col_min = df[col].min()
    col_max = df[col].max()

    # Normalization = x-min/max-min
    df[col] = (df[col] - col_min) / (col_max - col_min)
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Count the occurrences of each personality type
type_counts = df['Personality'].value_counts()

# Plot the bar chart
plt.bar(type_counts.index, type_counts.values, color='skyblue')
plt.xlabel('Personality Type')
plt.ylabel('Count')
plt.title('Distribution of Personality Types')
plt.xticks(rotation=45)  # Rotate x labels for readability if needed
plt.tight_layout()
plt.show()
```
